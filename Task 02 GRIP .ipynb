K-means Clustering on the Iris dataset

 Step 1: Importing the required libraries 

```python
#importing the libraries 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

Step 2: Uploading and Reading the dataset 

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

Step 3: Training the Kmeans model of dataset

```python
#Training the data and clustering using k=5

x = data.iloc[:, [0, 1, 2, 3]].values
```

```python
kmeans= KMeans(n_clusters=5)
y_kmeans=kmeans.fit_predict(x)
print(y_kmeans)
kmeans.cluster_centers_
```

Step 4: Using the Elnow method to find out the optimal number of clusters

```python
#Elbow method

Err=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init = 10, random_state = 0)
    kmeans.fit(x)
    Err.append(kmeans.inertia_)
```

Step 5: Plotting the data

```python
plt.plot(range(1,11),Err)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Err')
plt.show()
```

```python
#Clustering using k=3

kmeans = KMeans(n_clusters = 3, init = 'k-means++', n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
kmeans.cluster_centers_
```

Step 6: Visualising the data 

```python
# Visualising the clusters - On the first two columns

plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')
plt.legend()
plt.show()
```

```python
#Visualization of 3 Clusters
plt.scatter(x[:,0],x[:,1], c=y_kmeans,cmap='viridis')
plt.show()
```
